```python
# Operações Matemáticas Básicas

# Adição
print(5+23)

# Subtração
print(100-25)

# Multiplicação
print(5*10)

# Expoent
# ** operador é equivalente ao expoente
print(5**2)

# 5*5 = 5^2 = 5**2
print(5*5)


# Divisão (float)
# Voltar o valor decimal do valor de divisão
print(36/4)
print(10/3)

# Division (int)
# Voltar uma int. é dado o número inteiro
print(10//3)
print(19//6)

# Divisão modular: Volta o que sobrou da divisão
print(10%3)
```

```python
#operações com Strings e Caractéres
print ("foo" * 5)
print ("x"* 3)
print ("hello" + str (5))
print ("hello " + "world")
```

```python
# Comparadores: retornam um valor booleano

# Igualdade ==
# Nota: DEVEM ser usados dois sinais de igual,
# um único sinal de igual é atribuição
print(5 == 5.0)

# Maior que >
print(7 > 1)

# Menor que <
print(1.5 < 90)

# Maior ou igual a >=
print(5.0 >= 5)
print(5.0 >= 4)
print(5 >= 13)

# Menor ou igual a <=
print(10 <= 10.0)
print(10 <= 20)
print(8 <= 3)
```

```python
# Comparadores com Strings

print("hello" < "world")
print("hello" == "world")
print("hello" > "world")

print("hello" == "hello")

print("cat" < "dog")
```

```python
x = 2
y = 4

if (2*x == y):
  print("y é o dobro de x")
elif (x**2 == y):
  print("y é o quadrado de x")
else:
  print("y NÃO é o dobro de x")
```
